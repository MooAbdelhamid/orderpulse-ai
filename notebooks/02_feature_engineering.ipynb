# Feature Engineering

## 1. Setup

```python
import numpy as np
import pandas as pd
```

## 2. Load Cleaned Datasets

```python
daily = pd.read_csv('../data/processed/daily_clean.csv')
hourly = pd.read_csv('../data/processed/hourly_clean.csv')

daily["timestamp"] = pd.to_datetime(daily["timestamp"])
hourly["timestamp"] = pd.to_datetime(hourly["timestamp"])
```

## 3. Daily Feature Engineering

TIME FEATURES

```python
daily['day_of_week'] = daily['timestamp'].dt.day_of_week
```

LAG FEATURES

```python
daily["lag_1"] = daily["demand"].shift(1)
daily["lag_7"] = daily["demand"].shift(7)
```

ROLLING FEATURES

```python
daily["rolling_7"] = daily["demand"].rolling(7).mean()
```

```python
daily = daily.dropna()
```

## 4. Hourly Feature Engineering

TIME FEATURES

```python
hourly['hour'] = hourly['timestamp'].dt.hour
hourly['day_of_week'] = hourly['timestamp'].dt.dayofweek
hourly['is_weekend'] = hourly['day_of_week'].isin([5, 6]).astype(int)

hourly["hour_sin"] = np.sin(2 * np.pi * hourly["hour"] / 24)
hourly["hour_cos"] = np.cos(2 * np.pi * hourly["hour"] / 24)
```

LAG FEATURES

```python
hourly['lag_1'] = hourly['demand'].shift(1)
hourly['lag_24'] = hourly['demand'].shift(24)
```

ROLLING FEATURES

```python
hourly['rolling_24'] = hourly['demand'].rolling(24).mean()
```

```python
hourly = hourly.dropna()
```

## 5. Export Final Datasets

```python
daily.to_csv("../data/processed/daily_features.csv", index=False)
hourly.to_csv("../data/processed/hourly_features.csv", index=False)
```
